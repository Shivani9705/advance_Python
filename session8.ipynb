# index Error 

```python
lst=[1,2,3,4,6,7,10]
print(lst[10])
```

```python
try:
    lst=[10,20,33,44,12,23,34,45]
    indexnumber=int(input("Enter your index number"))
    print(lst[indexnumber])
except IndexError:
    print("List index out of range")
```

# Type error

```python
a='wdfghnm'
b=23
c=a+b 
print(c)
```

```python
try:
    a='wdfghnm'
    b=23
    c=a+b 
    print(c)
except TypeError as e:
    print(e)
```

# raise keyword 

```python
try:
    age=int(input("Enter your age"))
    if age<18:
        raise TypeError("Person is not eligible for voting")
    else:
        print("Person is eligible for voting ")
except TypeError as e:
    print(e)
```

# finally keyword

```python
try:
    a=int(input("Enter first number"))
    b=int(input("Enter second number"))
    c=a+b
    print("c=",c)
except ValueError as e:
    print(e)
finally:
    print("Finally your calculation is completed")
```
